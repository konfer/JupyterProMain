```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def showImg(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
def imread(image):
    image=cv2.imread(image)
    image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    return image
```

```python
image=imread('image.jpg')
```

```python
(R,G,B)=cv2.split(image)
```
